I use a dataset provided by scikit-learn and fit some models and provide data visualisation.

** I find that for pure regression analysis, statsmodel might be better suited, as it offers for example straight calculations of p-values and thus coefficients significance.**

```python
from sklearn import datasets 
import sklearn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
boston = datasets.load_boston()

# Data
X = pd.DataFrame(boston.data)
X.columns = boston.feature_names
Y = pd.DataFrame(boston.target)
Y.columns = ['MEDV']

# Description of the dataset
print(boston.DESCR)
```

```python
# Put it into one DF 
data = X
data['MEDV'] = Y['MEDV']
print(data.head())
print(data.describe())
```

```python
# Distribution plot 
%matplotlib notebook
sns.set(rc={'figure.figsize':(12,5)})
sns.distplot(data['MEDV'], bins=30)
```

```python
# Correlation matrix 
sns.set(style="white")

# Compute the correlation matrix
corr_matrix = data.corr().round(2)

# Correlation matrix
%matplotlib notebook
sns.heatmap(data=corr_matrix, annot=True) # annot = annotate number within squares
```

## Housing Price Analysis

* The housing prices seem to be normally distributed with few outliers. Linear regression might be appropriate. 
* Based on the correlation matrix: 
    - MEDV highly correlated INDUS, RM, TAX, PTRATIO, LSTAT
    - Since TAX and INDUS are highly correlated, I drop one of the two from the model (keep TAX)

```python
# Plot for these five variables 
%matplotlib notebook
plt.figure(figsize=(10, 8))

features = ['RM', 'TAX', 'PTRATIO', 'LSTAT']
target = data['MEDV']

for i, col in enumerate(features):
    plt.subplot(2, 2, i+1)
    plt.scatter(data[col], target, marker='o')
    plt.title(str(col) + ' vs MEDV')
```

```python
# Linear Regression 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X[features], Y, test_size = 0.2, random_state=5)
```

```python
# Fit linear regression
lm = LinearRegression()
lm.fit(X_train, Y_train)

# Predict 
Y_pred = lm.predict(X_test)

# Plot predictions
%matplotlib notebook
plt.figure(figsize=(10, 5))
plt.scatter(Y_test, Y_pred)
plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Prices vs Predicted prices: $Y_i$ vs $\hat{Y}_i$ (subset)")
```

```python
# Mean Squared errors 
def RMSE(errors): 
    # errors is an array of errors
    squared_errors = [e**2 for e in errors]
    return np.sqrt(np.mean(squared_errors))

errors = np.array(Y_test - Y_pred)
print('Root Mean Squared Error (RMSE) is: {}'.format(RMSE(errors).round(4)))
```

```python
# Coefficients Table
coefficients = pd.DataFrame(lm.coef_, columns = features)
coefficients['intercept'] = lm.intercept_

print(coefficients)
```

## Iris dataset - Classification 

*Classification using decision tree and KNN in scikit-learn.*

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
print(iris.DESCR)
data = iris.data
target = iris.target
```

```python
# DataFrame representation
df = pd.DataFrame(data)
df = pd.concat([df, pd.DataFrame(target)], axis = 1)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'cat']
print('Data Overview:\n')
print(df.head())
print('\nSummary:\n')
print(df.describe())
```

```python
# Plots
%matplotlib notebook
sns.set(rc={'figure.figsize':(10,12)})
plt.subplot(2,2,1)
sns.scatterplot(df.sepal_length, df.sepal_width, hue = df['cat']).set_title('Sepal')
plt.subplot(2,2,2)
sns.scatterplot(df.petal_length, df.petal_width, hue = df['cat']).set_title('Petal')
plt.subplot(2,2,3)
sns.scatterplot(df.petal_length, df.sepal_width, hue = df['cat']).set_title('Petal')
plt.subplot(2,2,4)
sns.scatterplot(df.sepal_length, df.petal_width, hue = df['cat']).set_title('Petal')
```

```python
# Training and Testing sets
from sklearn.model_selection import train_test_split
X = np.array(df.iloc[:,0:4])
Y = np.array(df.iloc[:, 4])

# Splitting the dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=5)
```

```python
# Fit Decision Tree
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, Y_train)

# Predict
predicts = clf.predict(X_test)

# Accuracy 
accuracy = [1 if ele == Y_test[idx] else 0 for idx, ele in enumerate(predicts) ]
accuracy = np.mean(accuracy)
print('The accuracy of Decision Tree is of:', np.round(accuracy*100, 2), '(%)')
```

```python
# Graphical Represenation 
import graphviz
dot_data = tree.export_graphviz(clf, out_file = None, 
                                feature_names = iris.feature_names, 
                                class_names = iris.target_names, 
                                filled = True, 
                                rounded = True, 
                                special_characters = True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
# Fit KNN 
from sklearn.neighbors import KNeighborsClassifier 
nbrs = KNeighborsClassifier(n_neighbors = 3)
nbrs = nbrs.fit(X_train, Y_train)

# Predict 
predicts_knn = nbrs.predict(X_test)
print(predicts_knn)
print(Y_test)

# Accuracy
accuracy_knn = [1 if ele == Y_test[idx] else 0 for idx, ele in enumerate(predicts_knn) ]
accuracy_knn = np.mean(accuracy_knn)
print('The accuracy of Decision Tree is of: {}%'.format(np.round(accuracy_knn*100, 2)))
```
